```python
import pandas as pd
path="C:\\Users\\Vishva\\OneDrive\\Desktop\\python\\Data\\ObesityDataSet_raw_and_data_sinthetic.csv"
data=pd.read_csv(path)
print(data)
print(data.info())

import sklearn 
from sklearn.preprocessing import LabelEncoder
le_Gender=LabelEncoder()
le_family_history_with_overweight=LabelEncoder()
le_FAVC=LabelEncoder()
le_CAEC=LabelEncoder()
le_SMOKE=LabelEncoder()
le_SCC=LabelEncoder()
le_CALC=LabelEncoder()
le_MTRANS=LabelEncoder()
le_NObeyesdad=LabelEncoder()


data['Gender_n']=le_Gender.fit_transform(data['Gender'])
data['family_history_with_overweight_n']=le_Gender.fit_transform(data['family_history_with_overweight'])
data['FAVC_n']=le_Gender.fit_transform(data['FAVC'])
data['CAEC_n']=le_Gender.fit_transform(data['CAEC'])
data['SMOKE_n']=le_Gender.fit_transform(data['SMOKE'])
data['SCC_n']=le_Gender.fit_transform(data['SCC'])
data['CALC_n']=le_Gender.fit_transform(data['CALC'])
data['MTRANS_n']=le_Gender.fit_transform(data['MTRANS'])
data['NObeyesdadr_n']=le_Gender.fit_transform(data['NObeyesdad'])

print(data)

inputs=data.drop(['Gender','family_history_with_overweight','FAVC','CAEC','SMOKE','SCC','CALC','MTRANS','NObeyesdad','NObeyesdadr_n'],'columns')
print(inputs)
outputs=data.drop(['Gender','family_history_with_overweight','FAVC','CAEC','SMOKE','SCC','CALC','MTRANS','NObeyesdad','Age','Height','Weight','FCVC','NCP','CH2O','FAF','TUE','Gender_n','family_history_with_overweight_n','FAVC_n','CAEC_n','SMOKE_n','SCC_n','CALC_n','MTRANS_n'],'columns')
print(outputs)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,train_size=0.8)


from sklearn.preprocessing import StandardScaler  # increase the accuracy
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.svm import SVC    # importing the model
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_test)
print(y_pred)

acc=model.score(x_train,y_train)*100
print("acc :",acc)



from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print("cm:",cm)


import numpy as np
newinputs = np.array([[21.000000,1.620000,64.000000,2.0,3.0,2.000000,0.000000,1.000000,0,1,0,2,0,0,3,3]])  
newinputs = sc.transform(newinputs)
res = model.predict(newinputs)
print("result:",res)
```
